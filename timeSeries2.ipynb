```python
import numpy as np
import pandas as pd
import datetime as dt
import re
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
random.randrange(1,100,1)
x = [x for x in range(0,11)]
y = [random.randrange(1,100,1) for x in range(0,11)]
```

```python
#figure is the larger background, the ax is a subplot on the figure
fig, ax = plt.subplots(figsize = (20,10))
#plot data on the subplot
ax.plot(x,y)

#use to force specific value ticks on the x-axis and at what intervals
ax.set_xticks([x for x in range(0,11,1)])
#sets the title name and font size
ax.set_title("Muh plot!",fontdict={'fontsize' : 18})

#get access to the individual "ticks" and set fontsize, fontname, and font colour by iterating through all tick labels
xAxes = ax.get_xticklabels()
yAxes = ax.get_yticklabels()
for z in xAxes+yAxes:
    z.set_fontsize(18)
    z.set_fontname('Arial')
    z.set_color('lightblue')

#get access to the plot lines and set the colour of the plot line
line = ax.get_lines()
line[0].set_color('yellow')

#set the colour of subplot to black, and also the figure to black
ax.set_facecolor('black')
fig.set_facecolor('black')

#change the colours of axis lines, and change their width
ax.spines['left'].set_color('lightblue')
ax.spines['bottom'].set_color('lightblue')
ax.spines['bottom'].set_linewidth(3)
ax.spines['left'].set_linewidth(3)

#add a random horizontal line
ax.hlines(70,0,10,colors='r')

#preset styles for plots
#plt.style.use('seaborn-pastel')

#after setting present style, I can continue editing other things like grid lines
ax.grid(True,linestyle='--')
#get all grid line objects into gridLines and iterate through that to change colour
gridLines = [t for t in (ax.get_xgridlines()+ax.get_ygridlines())]
for t in gridLines:
    t.set_color('dimgrey')
```
